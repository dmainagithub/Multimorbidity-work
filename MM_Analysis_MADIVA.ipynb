### **Import libraries**

```python
import pandas as pd
import numpy as np
```

### **1. Load data**

```python
df = pd.read_csv('/content/drive/MyDrive/MADIVA_DATA/DataSheet.csv')
df.head()
```

### **2. Handle Missing Values (string and values)**

```python
#-------------------------------------------------------------------------------
# Non communicable diseases
ncds = ['asthma', 'diabetes', 'epilepsy', 'hiv', 'hypertension', 'raised_bp', 'mental_illness', 'tb']
for col in ncds:
    df[col].fillna("", inplace=True)#-------------------------------------------------------------------------------
#-------------------------------------------------------------------------------
# Strings
string_columns = ['Sex', 'HealthFacility']
for col in string_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

#-------------------------------------------------------------------------------
#-------------------------------------------------------------------------------
#-------------------------------------------------------------------------------
#-------------------------------------------------------------------------------
```
